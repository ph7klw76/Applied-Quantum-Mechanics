```
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

length_x=10  # define the length of your box. you can change that
length_y=5    # define the width of your box. you can change that
x= np.arange(0, length_x, 0.1)
y= np.arange(0, length_y, 0.1)
x, y = np.meshgrid(x, y)
def wavefunction(Lx,Ly,nx,ny):
    wave = 2*np.sin(np.pi*x*nx/Lx)*np.sin(np.pi*y*ny/Ly)/np.sqrt(Lx*Ly)
    return wave
z = wavefunction(length_x,length_y,2,4)  # 1 is the nx and ny (the interger corresponding to the overtune of the vibration) You can change that
fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(x, y, z, rstride=1, cstride=1, cmap=cm.viridis)
plt.show()

fig, ax = plt.subplots()
CS = ax.contour(x, y, z)
ax.clabel(CS, inline=1, fontsize=10)
ax.set_title('Simplest default with labels')
plt.show()


fig, ax = plt.subplots()
im = ax.imshow(z, interpolation='bilinear', cmap=cm.RdYlGn,
               origin='lower', extent=[0, length_x, 0, length_y],
               vmax=abs(z).max(), vmin=-abs(z).max())

plt.show()
```
